## Overview
---
isomiR-SEA is an alignment tool for the analisys and quantification of small non-coding RNA and isoforms, written in modern C++.  

### Main features
- Highly customizable and flexible: you can set a large number of specific parameters.
- Compatible with [MirGeneDB](https://mirgenedb.org/) (GFF v2.0) and [miRBase](http://www.mirbase.org/) (GFF v22) reference database or with user-provided mature/star sequences.
- Support serialization: input data can be serialized to minimize the acquisition time of subsequent executions.
- Support multi-sample: you can provide tag-file containing collapsed sequences of different samples, like those produced by [BioSeqZip](https://github.com/bioinformatics-polito/BioSeqZip).
- Compatible with [Yara](https://github.com/seqan/seqan/blob/master/apps/yara/README.rst) and [BWA](https://github.com/bioinformatics-polito/BioSeqZip-BWA): for example you can map isomiR-SEA discarded or not alligned tags directly on precursors with Yara mapper.
- Very detailed and exhaustive alignment output (*.tab* and *.gff* format);  
- Computational costs contained despite the high degree of detail provided;  
- Support multi-species analysis.  
  

## Setup
---
This manual is intended for operating systems of the GNU / Linux family. However this does not prevent you from using the operating system you prefer.

### Software requirements
In order to build and run isomiR-SEA you need a modern C++ compiler (with [OpenMP](https://www.openmp.org/) extensions), [CMake](https://cmake.org/) tool and [Git](https://git-scm.com/).  
Furthermore this tool makes use of a modern C++ library for sequence analysis named Seqan. In particular isomiR-SEA uses both the current version of the library ([Seqan3](http://docs.seqan.de/seqan/3-master-user/index.html)) which is automatically included on the fly during compilation and the previous one ([Seqan2](https://seqan.readthedocs.io/en/master/index.html#)) which instead can be installed manually with a command below (or by following the tutorials on the [official website](https://seqan.readthedocs.io/en/master/Infrastructure/Use/index.html)).  
  
**gcc >= 7**  
`# sudo apt install g++`  
  
**cmake >= 3.4**  
`# sudo apt-get install cmake`  
  
**git**  
`# sudo apt install git`  
  
**Seqan2**  
`# sudo apt install libseqan2-dev`  
Note: this package is available in ubuntu since 18.04LTS, if you have an older version or another OS you can alternatively:  
-refer to SeqAn2 installation guide [here](https://seqan.readthedocs.io/en/master/Infrastructure/Use/Install.html);  
-download the latest version from [here](http://packages.seqan.de/), extract the content and copy the "/include/seqan" directory into the project "include" folder.   
  
**ZLIB, BZip2 and Boost libraries** (depending on your distribution you may also need to install these libraries)  
`# sudo apt install zlib1g-dev libbz2-dev`  
`# sudo apt-get install libboost-dev`
  
**Yara** (optional)  
If you want to align isomiR-SEA discarded reads directly on the miRNA precursors you need to install Yara tool following [this](https://github.com/seqan/seqan/blob/master/apps/yara/README.rst#installation-from-sources) tutorial and then proceed with reading this manual.

### Build
Create a build directory using CMake as follows:  
`# mkdir cmake-build-release`  
`# cd cmake-build-release`  
`# cmake ..`

Note: if you have installed g++(>=7) but your current default compiler is different, you can invoke cmake as follows:  
`# cmake -D CMAKE_CXX_COMPILER=g++-7 ..`
  
Invoke make as follows:  
`# make`  
  
You will find your executable here:  
*cmake-build-release/apps/__isomir_sea__*

## Input & Output


### Input files
You can provide different kind of input file to isomir-SEA:
- **File of precursors (option: --in-file-primir)**  
Precursor files are available online, you can alternatively download it from from MirGeneDB [download-page](https://mirgenedb.org/download) (precursor with 30nt-flank), from miRBase [download-page](http://www.mirbase.org/ftp.shtml) (hairpin file), or with wget command (see [Usage and configurations](#Usage-and-configurations)).  
Please note that *if you provide a precursor file you must also provide a genomic coordinates file.*  
  
  
- **File of genomic coordinates (option: --in-file-gff)**  
Genomic coordinates files are available online, you can alternatively download the GFF file (v2.0) from MirGeneDB [download-page](https://mirgenedb.org/download), all .gff3 files (v22) from miRBase [download-page](http://www.mirbase.org/ftp.shtml) (see [Merge GFF files](#Merge-GFF-files) section), or with wget command (see [Usage and configurations](#Usage-and-configurations)).  
Please note that *if you provide a genomic coordinates file you must also provide a precursor file.*
  
  
- **File of mature and/or star sequences (options: --in-file-mature, --in-file-star)**    
Mature and/or star sequences files are available online, from [MirGeneDB](https://mirgenedb.org/download) or [miRBase](http://www.mirbase.org/ftp.shtml) download-pages or with wget command (see [Usage and configurations](#Usage-and-configurations)), but you can also provide your own sequences only paying attention to stick to the format used in the files of the sites just mentioned.
  
  
- **Directly the serialized reference database (options: --store-serialized, --load-serialized)**  
You can run isomiR-SEA specifying a path where to save the internal datastructure filled with your input data (option: --store-serialized). Doing so, the next times you run the tool, you can directly load the serialized database (option: --load-serialized) instead of re-reading the same input files, saving execution time.  
  
  
- **File of tags/reads (options: --in-file-tags, --multi-sample-tab)**  
You can provide your own tags/reads file (.fasta, .fastq or .tag), even multi-sample.  
In this last case you will also provide the tabular file (option: --multi-sample-tab) containing informations about the number of reads per sample of each sequence. To produce this kind of collapsed file we used [BioSeqZip](https://github.com/bioinformatics-polito/BioSeqZip), an exact collapser for Second-Generation Sequencing datasets, so we kindly suggest you to use it.  
  
**Important**: see Use-Cases sections to understand the several possible input files combinations to use.

### Output files
isomiR-SEA generates three main output files: **.log, .tag, .gff**.  
Here the explanation of every single field/column name:  
- _tag_id_ (**TI**): an integer value that represents a unique index given to each tag while reading the input tags-counts file, an incremental number is given for each new row read in that file;  
- _tag_seq_ (**TS**): the tag sequence;  
- _tag_qual_ (**TQ**): the average quality (for each nucleotide) of the tag sequence;  
- _tag_count_ (**TC**): the number of identical reads of each tag; 
- _tag_start_gen_ (**TSG**): the start position (coordinate) of the tag into genome;  
- _tag_end_gen_ (**TEG**): the end position (coordinate) of the tag into genome;  
- _org_code_ (**ORG**): a three letters long identification code for the organism of the miRNA sequence;  
- _mir_id_ (**MI**): an integer value that represents a unique index given to each miRNA sequence while reading the input reference files, an incremental number is given for each new sequence;  
- _mir_seq_ (**MS**): the miRNA sequence which has been aligned to the tag;  
- _align_mark_ (**AM**): the alignment score;  
- _align_iupac_tag_ (**IT**): the iupac string related to the alignment with respect to the tag;  
- _align_iupac_mir_ (**IM**): the iupac string related to the alignment with respect to the miRNA;  
- _align_cigar_ (**CI**): the cigar string that is a compressed representation of the alignment;  
- _align_length_ (**AL**): the length of the alignment;  
- _mir_tag_size_diff_ (**SD**): the difference in length between the tag and the miRNA;  
- _iso_exact_ (**IEX**): a boolean indicating whether the sequence is identical the canonical mature sequence reported in the database;  
- _iso5p_ (**I5P**): an integer indicating whether the sequence is an isoform resulting from insertion or deletion in the 5p-end (**+**: insertion, **-**: deletion);  
- _iso_m_snp_ (**IMS**): stands for _multiple nucleotide polymorphism isoforms_ (iso_multi_snp) and it's a boolean indicating if the sequence presents more than a mismatch with respect to miRNA sequence;  
- _iso_snp_ (**ISN**): stands for _single nucleotide polymorphism isoforms_ and it's a boolean indicating if the sequence presents a mismatch with respect to miRNA sequence;  
- _iso3p_ (**I3P**): an integer indicating whether the sequence is an isoform resulting from insertion or deletion in the 3p-end (**+**: insertion, **-**: deletion);  
- _mismatch_in_seed_ (**INS**): a boolean indicating whether there are mismatches in the seed or not;  
- _off_site_ (**IOS**): a boolean indicating whether the offset site (nucleotide 8) is conserved in the tag sequence;  
- _suppl_site_ (**ISS**): a boolean indicating whether the supplementary site (from nucleotide 13 to 16) is conserved in the tag sequence;  
- _compens_site_ (**IPS**): a boolean indicating whether the compensatory site (from nucleotide 12 to 20 approximately) is conserved in the tag sequence;  
- _central_site_ (**ICS**): a boolean indicating whether the central site (from nucleotide 4 to 16 approximately) is conserved in the tag sequence;  
- _a2i_in_seed_ (**AIS**): stands for _adenine to inosine in seed_ and it's a boolean indicating if, inside the seed, an "A" in the miRNA correspond to a "G" in the tag;  
- _a2i_out_seed_ (**AIO**): stands for _adenine to inosine out seed_ and it's a boolean indicating if, outside the seed, an "A" in the miRNA correspond to a "G" in the tag;  
- _num_of_mir_same_seq_ (**MI4S**): an integer indicating the number of miRNAs the tag TI was aligned with;  
- _score_diff_mir0_mir1_ (**MSD**): an integer indicating the score difference between two subsequent alignment of miRNAs (same organism) with the same tag sequence (ex: between the score of the 3rd best aligned miRNA and the score of the 2nd one);  
- _mir_info_index_ (**MII**): an integer value that represents a unique index given to each miRNA sequence while reading the input reference files, an incremental number is given for each new sequence (it differs from _mir_id MI_ because identical miRNA sequences from different organisms have different _MII_ but same _MI_ );   
- _mir_info_info_ (**MIN**): a string indicating the full information of the miRNA sequence;  
- _mir_info_ref_ (**MRF**): a string indicating the reference chromosome of the miRNA;  
- _mir_info_start_gen_ (**MSG**): the start position (coordinate) of the miRNA into genome;  
- _mir_info_end_gen_ (**MEG**): the end position (coordinate) of the miRNA into genome;  
- _mir_info_mimat_ (**MIM**): the MIMAT (miRBase identifier) of the miRNA sequence;  
- _mir_info_strand_ (**MSR**): the strand of the miRNA sequence;  
- _prx_index_ (**PII**): an integer value that represents a unique index given to each precursor sequence while reading the input reference files, an incremental number is given for each new sequence;  
- _prx_info_ (**PIN**): a string indicating the full information of the precursor sequence;  
- _prx_ref_ (**PRF**): a string indicating the reference chromosome of the precursor sequence;  
- _prx_seq_ (**PS**): the precursor sequence from which the miRNA originates; 
- _prx_start_gen_ (**PSG**): the start position (coordinate) of the precursor into genome;  
- _prx_end_gen_ (**PEG**): the end position (coordinate) of the precursor into genome;   
- _prx_mimat_ (**PMI**): the MIMAT (miRBase identifier) of the precursor sequence;  
- _prx_strand_ (**PSR**): the strand of the precursor sequence;  
- _num_of_prx_same_seq_ (**PR4S**): an integer indicating the number of different precursors in which the miRNA sequence appears;  
- _prx_align_mark_ (**PAM**): the score of the alignment extended to the precursor (increased with each new match);  
- _iso5p_canonic_ (**IC5**): a boolean indicating whether the 5p end of the tag (insertion/s) is aligned on the precursor;  
- _iso3p_canonic_ (**IC3**): a boolean indicating whether the 3p end of the tag (insertion/s) is aligned on the precursor;  
- _score_diff_prx0_prx1_ (**PSD**): an integer indicating the score difference between two subsequent alignment of the tag with different precursor sequences;  
- _mir_tag_out_id_ (**MTOID**): an integer value that represents a unique index given to each miRNA-tag alignment in output, an incremental number is given for each new row;  
- _mir_matched_id_ (**MMI**): an integer value that represents a unique index given to each subsequent miRNA aligned with the same tag sequence, an incremental number is given for each different miRNA;  
- _mir_info_id_ (**MIID**): an integer value that represents a unique index given to each miRNA sequence while reading the input reference files, an incremental number is given for each new sequence;  
- _premir_id_ (**PID**): an integer value that represents a unique index given to each precursor sequence while reading the input reference files, an incremental number is given for each new sequence;  
- _align_iupac_prx_ (**IP**): the iupac string related to the alignment with respect to the precursor;  
- _single_multiple_discarded_ (**SMD**): an integer value indicating if the current row is a unique-mapping, multiple-mapping or a mapping to be discarded.
  
**_.log_ file**  
Contains the list of options and parameters used in the program execution followed by all the alignment features detected during the alignment of a Tag over miRs, and a small report that shows execution times in the various phases together with some information like the number of Tags used, discarded, aligned and the number of miRs aligned.  
  
**_.tag_ file**  
Each line of this file contains the information about a specific Tag-miR alignment. All these informations are tab-separated and the overall output is ordered by sequence alphabetically.  
  
**_.gff_ file**  
Each line of this file contains, similarly to the _.tab_ file, the information about a specific Tag-miR alignment but in GFF format. All these informations are tab-separated a part from the last column which is the compression of several fields together separated by a semicolon. The overall output is ordered by tag sequence alphabetically, and then by score in descending order. In particular:  
- Column 1: _mir_info_ref_ (**MRF**);  
- Column 2: name of miRNA database;  
- Column 3: type of the record;  
- Column 4: _tag_start_gen_ (**TSG**);    
- Column 5: _tag_end_gen_ (**TEG**);  
- Column 6: _align_mark_ (**AM**);  
- Column 7: _mir_info_strand_ (**MSR**);  
- Column 8: _prx_strand_ (**PSR**);  
- Column 9 is composed of several fields:  
    - _tag_id_ (**TI**);  
    - _tag_seq_ (**TS**);  
    - _tag_count_ (**TC**);  
    - _prx_info_ (**PIN**);  
    - _align_cigar_ (**CI**);  
    - _mir_info_info_ (**MIN**);  
    - **ISO** which is a concatenation of:  
        - _iso_exact_ (**IEX**);  
        - _iso5p_canonic_ (**IC5**) (optional);  
        - _iso5p_ (**I5P**);  
        - _iso_m_snp_ (**IMS**);  
        - _iso_snp_ (**ISN**);  
        - _iso3p_ (**I3P**);  
        - _iso3p_canonic_ (**IC3**) (optional);  
    - **INT** which is a concatenation of:
        - _mismatch_in_seed_ (**INS**);  
        - _off_site_ (**IOS**);  
        - _suppl_site_ (**ISS**);  
        - _compens_site_ (**IPS**);  
        - _central_site_ (**ICS**);  
    - **FILTER** which is a field that can be "Pass" or "NotPass". For instance if there are two or more different miRNAs belonging to the same organism/specie that have matched with the same tag, only one of them will be "Pass" and only if it has a strictly higher score than the others. On the other hand, if there are two equal miRNAs but belonging to different precursors (and therefore with the same score) and these scores are higher than the other miRNAs then they will both have "Pass".

### Merge GFF files  
Unfortunately in miRBase there is not a single genomic coordinate file containing all the species, instead there is a file for each one of them.  
To deal with this inconvenience this script has been provided: it takes a bunch of .gff3 files in input and generates a single compact file.

```python
import os

# Replace these paths with yours
dir_in = "./io_files_test/mirbase/gff"
path_out = "./io_files_test/mirbase/all_gff.gff3"

files = []
for file in os.listdir(dir_in):
    if file.endswith(".gff3"):
        files.append(os.path.join(dir_in, file))

with open(path_out, "w") as outfile:
    for x in files:
        with open(x) as f:
            for line in f:
                if not line.startswith("#"):
                    outfile.write(line)
print("Single GFF created.")
```

### Use Case 3: MirGeneDB precursors *and* mature/star sequences
In case you are using MirGeneDB precursors and genomic coordinates as input files for isomiR-SEA (Use Case 1, MirGeneDB section) you can also provide, in addition, a file of mature sequences and/or a file of star sequences (like Use Case 2).  
Note that in this way, the tool takes each mature/star sequence and checks the precursor file: if there are sequences that do not belong to any precursor, a "fictitious" precursor is created for them (5p + 3p filled with 'N'). If, on the other hand, they correspond to a precursor, the tool searches for the position of the sequence on the precursor and checks if the coordinates of the gff file are consistent, if they are not, they are updated with those that have been discovered.  
**This prevents and corrects any errors in the MirGeneDB database and allows you to add any of your personal sequences to the analysis.**  
The tag file is the BioSeqZip-collapsing of trimmed [ENCFF005EAG.fastq](https://www.encodeproject.org/files/ENCFF005EAG/).  
  
**Shell commands**

### Use Case 5: Alignment parameters
You can set many parameters to run the tool according to your preferences. Here are described some of the most important (verbosity and alignment), and then there is an example of execution with them.  
For more information and to find out other parameters please refer to the output of the _--help_ command.
The example shown here uses miRBase sequences and the tag file is the BioSeqZip-collapsing of trimmed [ENCFF005EAG.fastq](https://www.encodeproject.org/files/ENCFF005EAG/).

**-V, --verbose**  
0: no outputs;  
1: displays global statistics;  
2: displays extensive statistics for each batch of reads;  
3: debug output.  
  
**-n, --specie-codes**  
Specie codes to be evaluated during the alignment at the same time (usually only one is used).  
**-o, --min-size-aln-stp1**  
Minimum size of ungapped alignment, starting from the seed, extending the alignment.  
**-p, --min-align-score**  
Minimum alignment score for considering a tag expression of a miR.  
**-q, --seed-start**  
Start position of the seed.  
**-r, --seed-end**  
End position of the seed.  
**-s, --max-start-pos-tag**  
Max index in tag position for starting the seed alignment.  
**-t, --min-size-tag**  
Minimum size of tag to be considered for the alignment.  
**-u, --thr-select-tags**  
Threshold used to select select high quality multimapped tags.  
**-v, --mismatches-out-seed**  
Number of mismatches allowed between miRNA and tags.  
**-z, --mismatches-in-seed**  
Number of mismatches allowed between miRNA seed and tags.  
  
**Shell commands**  
`# ./cmake-build-release/apps/isomir_sea --in-file-primir primir_file --in-file-gff gff_file --in-file-tags tag_file --specie-codes hsa:mmu --min-size-aln-stp1 10 --min-align-score 7 --seed-start 1 --seed-end 6 --max-start-pos-tag 5 --min-size-tag 15 --thr-select-tags 12 --mismatches-out-seed 3 --mismatches-in-seed 1 --verbose 1`  
  
**Script**  


```python
import subprocess
import sys

# Replace these paths with yours
exe_file = "./cmake-build-release/apps/isomir_sea"
primir_file = "./io_files_test/mirbase/hairpin.fa" # miRBase
gff_file = "./io_files_test/mirbase/all_gff.gff3" # miRBase
tag_file = "./io_files_test/tag/ENCFF005EAG.collapsed.tag"

proc = subprocess.Popen([exe_file,
                         "--in-file-primir", primir_file,
                         "--in-file-gff", gff_file,
                         "--in-file-tags", tag_file,
                         "--specie-codes", "hsa:mmu",
                         "--min-size-aln-stp1", "10",
                         "--min-align-score", "7",
                         "--seed-start", "1",
                         "--seed-end", "6",
                         "--max-start-pos-tag", "5",
                         "--min-size-tag", "15",
                         "--thr-select-tags", "12",
                         "--mismatches-out-seed", "3",
                         "--mismatches-in-seed", "1",
                         "--verbose", "1"],
                         stdout=subprocess.PIPE, stderr=subprocess.PIPE)
for c in iter(lambda: proc.stderr.read(1), b''): # stderr(=debug_stream) is used for all program outputs
    sys.stdout.write(c)
for c in iter(lambda: proc.stdout.read(1), b''): # stdout used for "-h/--help"
    sys.stdout.write(c)
print("End.")
```

### Use Case 7: isomiR-SEA & Yara
isomiR-SEA tool performs some checks on tags length, discarding those that are too short or too long compared to the microRNA sequences used as reference. Moreover, during the alignment phase, the tool also discards all those tags for which no alignment has been found or for which the alignment has a too low score.  
You can map all these isomiR-SEA discarded or not aligned tags directly on precursors with Yara mapper.  
For sake of completeness it is reported here an exhaustive call to isomiR-SEA: multi-sample is enabled (like the previous configuration) and the path to a tabular file is provided.  
After that Yara is called.
  
**Shell commands**  
isomiR-SEA  
`# ./cmake-build-release/apps/isomir_sea --in-file-primir primir_file --in-file-gff gff_file --in-file-tags tag_file --multi-sample-tab tabular_file --path-discarded-tags path_discarded --specie-codes mmu --mismatches-in-seed 1 --verbose 1`  
  
Yara  
`# gzip -c precursor_file > REF.fasta.gz`  
`# gzip -c isomirsea_discarded_tags_file > READS.fastq.gz`  
`# yara_indexer REF.fasta.gz -o REF.index`  
`# yara_mapper REF.index READS.fastq.gz -o READS.bam`
  
or (if index already calculated)  
`# gzip -c isomirsea_discarded_tags_file > READS.fastq.gz`  
`# yara_mapper index_already_calculated READS.fastq.gz -o READS.bam`  


**Script**  
isomiR-SEA

```python
import subprocess
import sys

# Replace these paths with yours
exe_file = "./cmake-build-release/apps/isomir_sea"
primir_file = "./io_files_test/mirbase/hairpin.fa" # miRBase
gff_file = "./io_files_test/mirbase/all_gff.gff3" # miRBase
tag_file = "./io_files_test/tag/ENCFF005EAG_ENCFF008VOH.collapsed.fq"
tabular_file = "./io_files_test/tag/ENCFF005EAG_ENCFF008VOH.collapsed.tab"
path_discarded = "./io_files_test/tag/ENCFF005EAG_ENCFF008VOH.collapsed/discarded_tags.fq"

proc = subprocess.Popen([exe_file,
                         "--in-file-primir", primir_file,
                         "--in-file-gff", gff_file,
                         "--in-file-tags", tag_file,
                         "--multi-sample-tab", tabular_file,
                         "--path-discarded-tags", path_discarded,
                         "--specie-codes", "mmu",
                         "--mismatches-in-seed", "1",
                         "--verbose", "1"],
                         stdout=subprocess.PIPE, stderr=subprocess.PIPE)
for c in iter(lambda: proc.stderr.read(1), b''): # stderr(=debug_stream) is used for all program outputs
    sys.stdout.write(c)
for c in iter(lambda: proc.stdout.read(1), b''): # stdout used for "-h/--help"
    sys.stdout.write(c)
print("End.")
```

Yara  

```python
import gzip
import shutil
import os
import subprocess

# Replace these paths with yours
out_dir = "./io_files_test/tag/ENCFF005EAG_ENCFF008VOH.collapsed/yara"
discarded_tags = "./io_files_test/tag/ENCFF005EAG_ENCFF008VOH.collapsed/discarded_tags.fq" # IsomiR-SEA output
primirs = "./io_files_test/mirbase/hairpin.fa" # miRBase

print("Creating .gz file...")
if not os.path.exists(out_dir):
    os.mkdir(out_dir)
with open(discarded_tags, 'rb') as f_tag_in, gzip.open(out_dir + "/READS.fastq.gz", 'wb') as f_tag_out:
    shutil.copyfileobj(f_tag_in, f_tag_out)
with open(primirs, 'rb') as f_ref_in, gzip.open(out_dir + "/REF.fasta.gz", 'wb') as f_ref_out:
    shutil.copyfileobj(f_ref_in, f_ref_out)
print("Creating index file...")
proc1 = subprocess.Popen(["yara_indexer", out_dir + "/REF.fasta.gz", "-o", out_dir + "/REF.index"])
proc1.wait()
print("Mapping...")
proc2 = subprocess.Popen(["yara_mapper", out_dir + "/REF.index", out_dir + "/READS.fastq.gz",
                          "-o", out_dir + "/READS.sam"])
proc2.wait()
print("End.")
```
